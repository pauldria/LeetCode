```python
def minStartValue(nums):
    rolling_sum = 0
    min_rolling_sum = 0 
    for n in nums:
        rolling_sum += n
        if rolling_sum < min_rolling_sum:
            min_rolling_sum = rolling_sum
    return -1*min_rolling_sum + 1
```

https://leetcode.com/contest/biweekly-contest-24/problems/find-the-minimum-number-of-fibonacci-numbers-whose-sum-is-k/

```python
def fib(n):
    if n < 1:
        return 0
    if n == 1 or n == 2:
        return 1
    f1 = 1
    f2 = 1
    f = f1 + f2
    for i in range(4, n+1):
        f1 = f2
        f2 = f
        f = f1 + f2
    return f
```

```python
fib(10)
```

```python
def findLargestFibonacciLessThanK(k):
    idx = 1
    f = fib(idx)
    old_f = f
    while f <= k:
        idx += 1
        old_f = f
        f = fib(idx)
    return old_f
```

```python
findLargestFibonacciLessThanK(55)
```

```python
def findMinFibonacciNumbers(k):
    cnt = 0
    sum = k
    while sum > 0:
        cnt += 1
        sum -= findLargestFibonacciLessThanK(sum)
    return cnt
```

```python
findMinFibonacciNumbers(10)
```

https://leetcode.com/contest/biweekly-contest-24/problems/the-k-th-lexicographical-string-of-all-happy-strings-of-length-n/

```python
def convertToString(n, k):
    m = {0:'a', 1:'b', 2:'c'}
    s = ""
    if n < 0:
        return ""
    if n == 0:
        return ('a'*k)
    while n > 0:
        div, mod = divmod(n, 3)
        n = div
        s = s + m[mod]
    # Prepend with a's
    return ('a'*(k-len(s)) + s[::-1])
def isGood(s):
    n = len(s)
    if n <= 1:
        return True
    for i in range(n-1):
        if s[i] == s[i+1]:
            return False
    return True
```

```python
[convertToString(i,2) for i in range(3**2)]
```

```python
[isGood(convertToString(i, 2)) for i in range(3**2)]
```

```python
def getHappyString(n, k):
    start = 0
    end = 3**n
    numgood = 0
    for i in range(start, end):
        s = convertToString(i, n)
        if isGood(s):
            numgood += 1
        if numgood == k:
            return s
    return ""
```

```python
getHappyString(10, 100)
```

https://leetcode.com/contest/biweekly-contest-24/problems/restore-the-array/

```python
saved = {}

def getCachedValue(orig_s, l, k):
    if orig_s in saved:
        if k in saved[orig_s]:
            return saved[orig_s][k][l]
        else:
            saved[orig_s][k] = [-1]*(len(orig_s)+1)
            return -1
    else:
        saved[orig_s] = {}
        saved[orig_s][k] = [-1]*(len(orig_s)+1)
        return -1

def numberOfArraysHelper(orig_s, l, k):
    cached_value = getCachedValue(orig_s, l, k)
    if cached_value > -1:
        return cached_value
    n_orig = len(orig_s)
    s = orig_s[n_orig-l:]
    n = len(s)
    if n == 0:
        return 0
    if s[0] == '0':
        return 0
    n_k = len(str(k))
    total = 0
    for i in range(1, n_k+1):
        if i < n:
            potential = s[:i]
            rest = s[i:]
            if int(potential) <= k:
                total += numberOfArraysHelper(orig_s, len(rest), k)
        if i == n:
            total += 1 if int(s) <= k else 0

    result = total % (10**9 + 7)
    saved[orig_s][k][l] = result
    return result

def numberOfArrays(s, k):
    total = numberOfArraysHelper(s, len(s), k)
            
    return (total % (10**9 + 7))
```
